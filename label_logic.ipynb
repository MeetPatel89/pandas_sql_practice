```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(data={"Price": [5, 100, 25, 20, 75, 125]})
df
```

```python
df.loc[df["Price"] < 30, "Category"] = "Low"
df.loc[(df["Price"] >= 30) & (df["Price"] <= 70), "Category"] = "Medium"
df.loc[df["Price"] > 70, "Category"] = "High"
```

```python
df
```

```python
df = pd.DataFrame(data={"Price": [5, 100, 25, 20, 75, 125]})
df
```

```python
df["Category"] = np.select(
    [
        df["Price"] < 30,
        (df["Price"] >= 30) & (df["Price"] <= 70),
        df["Price"] > 70
    ],
    [
        "Low",
        "Medium",
        "High"
    ]
)
```

```python
df
```

```python
df = pd.DataFrame(data={"Price": [5, 100, 25, 20, 75, 125]})
df
```

```python
def assign_label(x):
    if x < 30:
        return "Low"
    elif x >= 30 and x <= 70:
        return "Medium"
    elif x > 70:
        return "High"

df["Category"] = df["Price"].transform(assign_label)
df
```
